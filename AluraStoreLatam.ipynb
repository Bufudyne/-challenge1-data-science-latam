### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.tail()
```

#1. Análisis de facturación



```python
ingreso_total_tienda1 = tienda['Precio'].sum()
ingreso_total_tienda2 = tienda2['Precio'].sum()
ingreso_total_tienda3 = tienda3['Precio'].sum()
ingreso_total_tienda4 = tienda4['Precio'].sum()
```

```python
print(f"Ingreso total Tienda 1: {ingreso_total_tienda1}")
print(f"Ingreso total Tienda 2: {ingreso_total_tienda2}")
print(f"Ingreso total Tienda 3: {ingreso_total_tienda3}")
print(f"Ingreso total Tienda 4: {ingreso_total_tienda4}")
```

# 2. Ventas por categoría

```python
ventas_por_categoria_tienda1 = tienda.groupby('Categoría del Producto').size().sort_values(ascending=False)
ventas_por_categoria_tienda2 = tienda2.groupby('Categoría del Producto').size().sort_values(ascending=False)
ventas_por_categoria_tienda3 = tienda3.groupby('Categoría del Producto').size().sort_values(ascending=False)
ventas_por_categoria_tienda4 = tienda4.groupby('Categoría del Producto').size().sort_values(ascending=False)
print("Ventas por categoría Tienda 1:")
print(ventas_por_categoria_tienda1)
print("\n")
print("Ventas por categoría Tienda 2:")
print(ventas_por_categoria_tienda2)
print("\n")
print("Ventas por categoría Tienda 3:")
print(ventas_por_categoria_tienda3)
print("\n")
print("Ventas por categoría Tienda 4:")
print(ventas_por_categoria_tienda4)
```

# 3. Calificación promedio de la tienda


```python
calificacion_tienda1 = tienda['Calificación'].mean()
calificacion_tienda2 = tienda2['Calificación'].mean()
calificacion_tienda3 = tienda3['Calificación'].mean()
calificacion_tienda4 = tienda4['Calificación'].mean()
```

```python
print("Calificacion promedio Tienda 1:")
print(calificacion_tienda1)
print("\n")
print("Calificacion promedio Tienda 2:")
print(calificacion_tienda2)
print("\n")
print("Calificacion promedio Tienda 3:")
print(calificacion_tienda3)
print("\n")
print("Calificacion promedio Tienda 4:")
print(calificacion_tienda4)
```

# 4. Productos más y menos vendidos

```python
productos_tienda1 = tienda['Producto'].value_counts()

mas_vendidos_tienda1 = productos_tienda1.head(5)
menos_vendidos_tienda1 = productos_tienda1.tail(5)

print("Mas vendidos en Tienda 1:")
print(mas_vendidos_tienda1)
print("\nMenos Vendidos en Tienda 1:")
print(menos_vendidos_tienda1)

productos_tienda2 = tienda2['Producto'].value_counts()

mas_vendidos_tienda2 = productos_tienda2.head(5)
menos_vendidos_tienda2 = productos_tienda2.tail(5)

print("\nMas vendidos en Tienda 2:")
print(mas_vendidos_tienda2)
print("\nMenos Vendidos en Tienda 2:")
print(menos_vendidos_tienda2)

productos_tienda3 = tienda3['Producto'].value_counts()

mas_vendidos_tienda3 = productos_tienda3.head(5)
menos_vendidos_tienda3 = productos_tienda3.tail(5)

print("\nMas vendidos en Tienda 3:")
print(mas_vendidos_tienda3)
print("\nMenos Vendidos en Tienda 3:")
print(menos_vendidos_tienda3)

productos_tienda4 = tienda4['Producto'].value_counts()

mas_vendidos_tienda4 = productos_tienda4.head(5)
menos_vendidos_tienda4 = productos_tienda4.tail(5)

print("\nMas vendidos en Tienda 4:")
print(mas_vendidos_tienda4)
print("\nMenos Vendidos en Tienda 4:")
print(menos_vendidos_tienda4)
```

# 5. Envío promedio por tienda

```python
costo_envio_promedio_tienda1 = tienda['Costo de envío'].mean()
costo_envio_promedio_tienda2 = tienda2['Costo de envío'].mean()
costo_envio_promedio_tienda3 = tienda3['Costo de envío'].mean()
costo_envio_promedio_tienda4 = tienda4['Costo de envío'].mean()
```

```python
print("Costo de envío promedio Tienda 1")
print(costo_envio_promedio_tienda1)
print("\nCosto de envío promedio Tienda 2")
print(costo_envio_promedio_tienda2)
print("\nCosto de envío promedio Tienda 3")
print(costo_envio_promedio_tienda3)
print("\nCosto de envío promedio Tienda 4")
print(costo_envio_promedio_tienda4)
```

# 6. Visualizaciones

```python
import matplotlib.pyplot as plt

# Crear un diccionario con los ingresos totales de cada tienda
ingresos_totales = {
    'Tienda 1': ingreso_total_tienda1,
    'Tienda 2': ingreso_total_tienda2,
    'Tienda 3': ingreso_total_tienda3,
    'Tienda 4': ingreso_total_tienda4
}

# Convertir el diccionario a un DataFrame de pandas
df_ingresos = pd.DataFrame(list(ingresos_totales.items()), columns=['Tienda', 'Ingreso Total'])

# Generar el gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(df_ingresos['Tienda'], df_ingresos['Ingreso Total'], color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
plt.title('Ingreso Total por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Ingreso Total')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Consolidar los datos de ventas por categoría para todas las tiendas
df_ventas_categoria = pd.DataFrame({
    'Tienda 1': ventas_por_categoria_tienda1,
    'Tienda 2': ventas_por_categoria_tienda2,
    'Tienda 3': ventas_por_categoria_tienda3,
    'Tienda 4': ventas_por_categoria_tienda4
}).reset_index()
df_ventas_categoria.rename(columns={'index': 'Categoría del Producto'}, inplace=True)

# Convertir el DataFrame a formato largo para Seaborn
df_ventas_categoria_melted = df_ventas_categoria.melt(id_vars=['Categoría del Producto'], var_name='Tienda', value_name='Ventas')

# Crear el gráfico de barras agrupadas
plt.figure(figsize=(14, 8))
sns.barplot(x='Categoría del Producto', y='Ventas', hue='Tienda', data=df_ventas_categoria_melted, palette='viridis')

# Añadir título y etiquetas
plt.title('Ventas por Categoría de Producto por Tienda', fontsize=16)
plt.xlabel('Categoría del Producto', fontsize=12)
plt.ylabel('Número de Ventas', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Tienda')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
